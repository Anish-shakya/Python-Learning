### Using Logical operator to combine conditions to filter the data set

```python
is_longer_duration = df['duration'] > 90
is_action_movie = df['genre'] == 'Horror Movies'

df[is_longer_duration & is_action_movie]
```

#### Subsetting the dataframe using isin() method

```python
is_action_or_horror = df['genre'].isin(['Action','Horror Movies'])
df[is_action_or_horror]
```

## New Columns

```python

# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')

# Extract day, month, and year
df['day'] = df['date_added'].dt.day
df['month'] = df['date_added'].dt.month_name()
df['year'] = df['date_added'].dt.year

df
```

# Counting in Pandas

```python

Movie_Count = df["title"].value_counts()
print(Movie_Count)


Movie_Count = df["title"].value_counts(normalize=True)
print(Movie_Count)


Movie_Count = df["title"].value_counts(sort=True)
print(Movie_Count)

Movie_Count = df["title"].value_counts(sort=True,normalize=True)
print(Movie_Count)
```

```python
import numpy as np
df_pivot = df.pivot_table(values='duration', index='genre') ## by default it calculated the mean aggregation 
df_pivot

df_pivot_agg = df.pivot_table(values='duration',index='genre', aggfunc=[np.mean,np.median,np.max,np.min])
df_pivot_agg
```
